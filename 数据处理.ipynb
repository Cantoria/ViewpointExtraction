```python
import pandas as pd
comment_corpus = pd.read_csv("./dataset/网贷平台评论.csv",header=0, sep=",").drop_duplicates(subset=["评论内容"])
comment_corpus = comment_corpus.sort_values(by="平台名称" , ascending=False)
```

```python
'''
    NLP工具
'''
import pyltp 
from pyltp import NamedEntityRecognizer
from pyltp import Postagger
from pyltp import Segmentor
import os

LTP_DATA_DIR = "/home/shenyinghan/Project/Golaxy/model/ltp_data_v3.4.0"
# 分词
cws_model_path = os.path.join(LTP_DATA_DIR, 'cws.model')
segmentor = Segmentor()
segmentor.load(cws_model_path)
def _seg(x, seg, key):
    '''
        将句子分词
    '''
    return list(seg.segment(x[key]))
comment_corpus['分词'] = comment_corpus.apply(_seg, seg=segmentor, key='评论内容', axis=1)
```

```python
comment_corpus
```

```python
def _pos(x, pos):
    return list(pos.postag(x['分词']))
pos_model_path = os.path.join(LTP_DATA_DIR, 'pos.model')
postagger = Postagger()
postagger.load(pos_model_path)
comment_corpus['词性标注'] = comment_corpus.apply(_pos, pos=postagger, axis=1)

# 命名实体识别
def _rec(x, res):
    return list(res.recognize(x['分词'], x['词性标注']))
recognizer = NamedEntityRecognizer()
ner_model_path = os.path.join(LTP_DATA_DIR, 'ner.model')
recognizer.load(ner_model_path)
comment_corpus['命名实体识别'] = comment_corpus.apply(_rec, res=recognizer, axis=1)
```
